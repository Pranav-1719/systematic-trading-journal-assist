```python
#Imports
from datetime import date
```

```python
#Trades Class
class Trade:
    def __init__(self, symbol,trade_date):
        self.trade_date = trade_date
        self.symbol = symbol
        self.executed = False
        self.result = None

    def executed(self):
        self.executed = True

#Trades List
trades=[]

#Add Trade Function
def add_trade():
    #Input date and symbol
    trade_date= date.fromisoformat(input("Enter the date of the trade (YYYY-MM-DD): "))
    symbol = input("Enter the symbol of the trade: ")

    #Creates new trade object
    new_trade = Trade(trade_date,symbol)

    #adds object to list
    trades.append(new_trade)

    #Returns message
    print(f"{symbol} Trade added successfully on {trade_date}")


#View all Trades
def view_trades():
     #Returns the list of trades
      if len(trades)==0:
        print("\nNo Trades Available\n")

      else:
          print("\nAll Trades\n")
          for index,t in enumerate(trades):
              print(f"{index+1}.{t.trade_date} - {t.symbol} - {t.executed}\n")


#Mark trades as Executes
def mark_executed():
      view_trades()

      #Find Index
      trade_no=int(input("Enter the trade number to Execute: "))
      index=trade_no-1

      #Trade execution
      if 0<=trade_no<len(trades):
          trades[index].executed()
          print("Trade Executed Successfully")
      else:
        print("Trade not available!")

#Delete Trades
def delete_trade():
    view_trades()

    #Find Index
    trade_no=int(input("Enter the trade number to Delete: "))
    index=trade_no-1

    #Delete Trade
    if 0<=trade_no<len(trades):
        trades.pop(index)
        print("Trade Deleted Successfully")

    else:
        print("Trade not Available!")



# {Trade Dashboard
# if len(trades)==0:
#         print("No Trades Available")
#
#     else:
#         #returns Executed Trades
#         print("\nExecuted Trades\n")
#         for index,t in enumerate(trades):
#             if t.executed:
#                 print(f"{index+1}.{t.trade_date} - {t.symbol} - {t.executed} - {t.result}\n")
#
#         #Returns Pending Trades
#         print("\nPending Trades\n")
#         for index,t in enumerate(trades):
#             if not t.executed:
#                 print(f"{index+1}.{t.trade_date} - {t.symbol} - {t.result} \n")}
```
