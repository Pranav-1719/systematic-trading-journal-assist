```python
#Imports
from datetime import date
```

```python
#Trades Class
class Trade:
    def __init__(self, symbol,trade_date):
        self.trade_date = trade_date
        self.symbol = symbol
        self.executed = False
        self.result = None

    def set_executed(self):
        self.executed = True

    def set_result(self, pnl):
        self.result = pnl

#Trades List
trades=[]

#Add Trade Function
def add_trade():
    #Input date and symbol
    try:
        trade_date = date.fromisoformat(input("Enter the date of the trade (YYYY-MM-DD): "))
    except ValueError:
        print("Invalid date format. Please use YYYY-MM-DD.")
        return

    symbol = input("Enter the symbol of the trade: ")

    #Creates new trade object
    new_trade = Trade(symbol=symbol,trade_date=trade_date)

    #adds object to list
    trades.append(new_trade)

    #Returns message
    print(f"{symbol} Trade added successfully on {trade_date}")


#View all Trades
def view_trades():
     #Returns the list of trades
      if len(trades)==0:
        print("\nNo Trades Available\n")
        return


      print("\nAll Trades\n")
      for index,t in enumerate(trades):
        print(f"{index+1}.{t.trade_date} - {t.symbol} - {t.executed} - {t.result}")


#Mark trades as Executes
def mark_executed():

      if len(trades)==0:
        print("No Trades Available")
        return

      view_trades()

      #Find Index
      try:
          trade_no = int(input("Enter the trade number to Execute: "))

      except ValueError:
          print("Invalid input! Please enter a number.")
          return

      index=trade_no-1

      #Trade execution
      if 0<= index <len(trades):
          trades[index].set_executed()
          print("Trade Executed Successfully")
      else:
        print("Trade not available!")


#Delete Trades
def delete_trade():

    if len(trades)==0:
        print("No Trades Available")
        return

    view_trades()

    #Find Index
    try:
        trade_no = int(input("Enter the trade number to Delete: "))

    except ValueError:
        print("Invalid input! Please enter a number.")
        return

    index=trade_no-1

    #Delete Trade
    if 0<= index <len(trades):
        trades.pop(index)
        print("Trade Deleted Successfully")

    else:
        print("Trade not Available!")

#Calculate total P&L
def calculate_pnl():
    total_pnl = 0
    if len(trades)==0:
        print("No trades available to calculate P&L.")
        return
    else:
        for t in trades:
            if t.executed and t.result is not None:
                total_pnl += t.result


    print(f"Total Profit & Loss: {total_pnl}\n\n")

#Dashboard view
def view_dashboard():
    if len(trades)==0:
        print("No Trades Available")

    else:
        print("Welcome To Dashboard\n\n")

        calculate_pnl()

        #returns Executed Trades
        print("\nExecuted Trades\n")
        count=1
        for index,t in enumerate(trades):
            if t.executed:
                print(f"{count}.{t.trade_date} - {t.symbol} - {t.executed} - {t.result}")
                count+=1

        #Returns Pending Trades
        print("\nPending Trades\n")
        count=1
        for t in trades:
            if not t.executed:
                print(f"{count}.{t.trade_date} - {t.symbol} - {t.executed} - {t.result}")
                count+=1


#Update Profit and Loss
def update_trade():
    if len(trades)==0:
        print("No Trades Available")
        return

    view_trades()

    #Find Index
    try:
        choice = int(input("Enter the trade number to be Updated: "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        return


    if 0< choice <=len(trades):
         index=choice-1

    else:
         print("Enter valid Choice")
         return

    t=trades[index]

    #Update the P&L
    if t.executed:
        #Take the profit or Loss
        try:
            pnl = int(input("Enter the Profit or Loss: "))

        except ValueError:
            print("Invalid input! Please enter a number.")
            return

        t.set_result(pnl)
        print("Trade Updated Successfully")

    else:
        print("Trade not Executed!")


```
