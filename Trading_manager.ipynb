```python
#prerequisite
prerequisites={
     1: "Have I checked the fundamentals?",
    2: "Have I removed all unnecessary drawings from the chart?",
    3: "Have I marked the previous day‚Äôs high and low?",
    4: "On the 4-hour timeframe, have I identified the market direction and key swing levels?",
    5: "On the 4-hour timeframe, have I marked supply and demand zones along with the golden pocket?",
    6: "On the 1-hour timeframe, have I drawn trendlines and key zones?",
    7: "On the 1-hour timeframe, have I identified the market structure (choppy or trending)?",
    8: "On the 30-minute timeframe, have I spotted candlesticks and candlestick patterns?",
    9: "On the 15-minute timeframe, have I spotted candlesticks and candlestick patterns?"
}


#Rules for bullish trade
bullish_rules = {
    1: "Are the fundamentals bullish?",
    2: "Is the daily trend bullish?",
    3: "Is the 4-hour market structure bullish (trendy)?",
    4: "Is the 4-hour price at key levels or within a demand zone?",
    5: "Is the 4-hour stochastic oversold (below 20) and turning upward?",
    6: "Is the 1-hour trend bullish with price respecting the trendline?",
    7: "Is the 1-hour stochastic oversold (below 20) and turning upward?",
    8: "Have I spotted bullish candlesticks or candlestick patterns on the 30-minute timeframe?",
    9: "Is the 30-minute stochastic oversold (below 20) and turning upward?",
    10: "Have I spotted 30-minute candlestick patterns with long lower wicks?",
    11: "Is the 15-minute stochastic oversold (below 20) and turning upward?",
    12: "Has the price closed above the previous high for confirmation?",
    13: "Is the stop loss placed below a valid recent low?"
}


#Rules for bearish trade
bearish_rules = {
    1: "Are the fundamentals bearish?",
    2: "Is the daily trend bearish?",
    3: "Is the 4-hour market structure bearish (trendy)?",
    4: "Is the 4-hour price at key levels or within a supply zone?",
    5: "Is the 4-hour stochastic overbought (above 80) and turning downward?",
    6: "Is the 1-hour trend bearish with price respecting the trendline?",
    7: "Is the 1-hour stochastic overbought (above 80) and turning downward?",
    8: "Have I spotted bearish candlesticks or candlestick patterns on the 30-minute timeframe?",
    9: "Is the 30-minute stochastic overbought (above 80) and turning downward?",
    10: "Have I spotted 30-minute candlestick patterns with long upper wicks?",
    11: "Is the 15-minute stochastic overbought (above 80) and turning downward?",
    12: "Has the price closed below the previous low for confirmation?",
    13: "Is the stop loss placed above a valid recent high?"
}
```

<h3>Functions

```python
#Imports
from datetime import date



#Trades Class
class Trade:
    def __init__(self, symbol,trade_date,trend):
        self.trade_date = trade_date
        self.symbol = symbol
        self.trend = trend
        self.executed = False
        self.result = None

    def set_executed(self):
        self.executed = True

    def set_result(self, pnl):
        self.result = pnl


#Trades List
trades=[]


#Add Trade Function
def add_trade():
    #Input date and symbol
    try:
        trade_date = date.fromisoformat(input("üìÖ Enter the date of the trade (YYYYMMDD): "))
    except ValueError:
        print("‚ùå Invalid date format. Please use YYYY-MM-DD.")
        return

    symbol = input("üü¢ Enter the symbol of the trade: ")

    trend = input("üü¢ Enter the trend(Bullish / Bearish): ").strip().lower()

    if trend not in ("bullish", "bearish"):
        print("‚ùå Invalid trend value\n")
        return

    #Creates new trade object
    new_trade = Trade(symbol=symbol,trade_date=trade_date,trend=trend)

    #adds object to list
    trades.append(new_trade)

    #Returns message
    print(f"‚úÖ {symbol} - {trend} Trade added successfully on {trade_date}\n")

    if trend == "bullish":
        print("üü¢ Executing BUY logic")
        trade_probability(new_trade, trend)
    else:
        print("üü¢ Executing SELL logic")
        trade_probability(new_trade, trend)




#Trade probability calculator
def trade_probability(new_trade,trend):
    count=0

    #Bullish trend quiz
    if trend == "bullish":
        ruleset=bullish_rules

    #Bearish trend quiz
    else:
        ruleset=bearish_rules


    for rule in ruleset.values():
            response=input(f"{rule} (y/n): ").strip().lower()
            if response=="y":
                count+=1

    total_rules =len(ruleset)

    #Trade Probability calculation
    probability=count/total_rules

    if probability>=0.8:
        print(f"üü¢ The chances are {probability*100}% ,You can enter the Trade. ‚úÖ")

    elif 0.8>probability>0.5:
        print(f"üü¢ The chances are {probability*100}% ,Wait for a while. üïê ")

    elif probability<=0.5:
        print(f"üü¢ The chances are {probability*100}% ,Dont enter the Trade. ‚ùå ")

    else:
        print("‚ùå Invalid probability value")
        return

    final_step=input("üü¢ Did you take the trade (y/n): ").strip().lower()

    if final_step=="y":
        new_trade.set_executed()

    else:
        return


#View all Trades
def view_trades():
     #Returns the list of trades
      if len(trades)==0:
        print("\n‚ùå No Trades Available\n")
        return


      print("\nüü¢ ALL TRADES üü¢\n")
      for index,t in enumerate(trades):
        print(f"{index+1}.{t.trade_date} - {t.symbol} - {t.trend} -  Executed : {t.executed} - PnL: {t.result}")


#Mark trades as Executes
def mark_executed():

      if len(trades)==0:
        print("‚ùå No Trades Available")
        return

      view_trades()

      #Find Index
      try:
          trade_no = int(input("Enter the trade number to Execute: "))

      except ValueError:
          print("‚ùå Invalid input! Please enter a number.")
          return

      index=trade_no-1

      #Trade execution
      if 0<= index <len(trades):
          trades[index].set_executed()
          print("\n‚úÖ Trade Executed Successfully")
      else:
        print("\n‚ùå Trade not available!")


#Delete Trades
def delete_trade():

    if len(trades)==0:
        print("‚ùå No Trades Available")
        return

    view_trades()

    #Find Index
    try:
        trade_no = int(input("Enter the trade number to Delete: "))

    except ValueError:
        print("‚ùå Invalid input! Please enter a number.")
        return

    index=trade_no-1

    #Delete Trade
    if 0<= index <len(trades):
        trades.pop(index)
        print("üóëÔ∏è Trade Deleted Successfully")

    else:
        print("‚ùå Trade not Available!")


#Calculate total P&L
def calculate_pnl():
    total_pnl = 0
    if len(trades)==0:
        print("‚ùå No trades available to calculate P&L.")
        return
    else:
        for t in trades:
            if t.executed and t.result is not None:
                total_pnl += t.result


    print(f"\nüíπ Total Profit & Loss: {total_pnl} üíπ\n")


#Dashboard view
def view_dashboard():
    if len(trades)==0:
        print("‚ùå No Trades Available")

    else:
        print("\nüéØ WELCOME TO DASHBOARD üéØ\n")

        calculate_pnl()

        #returns Executed Trades
        print("\n‚úÖ Executed Trades\n")
        count=1
        for index,t in enumerate(trades):
            if t.executed:
                print(f"{count}.{t.trade_date} - {t.symbol} - {t.trend} -  Executed : {t.executed} - PnL: {t.result}")
                count+=1

        #Returns Pending Trades
        print("\nüïê Pending Trades\n")
        count=1
        for t in trades:
            if not t.executed:
                print(f"{count}.{t.trade_date} - {t.symbol} - {t.trend} - Executed : {t.executed} - PnL: {t.result}")
                count+=1

    print("\n\n")


#Update Profit and Loss
def update_trade():
    if len(trades)==0:
        print("‚ùå No Trades Available\n")
        return

    view_trades()

    #Find Index
    try:
        choice = int(input("Enter the trade number to be Updated: "))
    except ValueError:
        print("‚ùå Invalid input! Please enter a number.\n")
        return


    if 0< choice <=len(trades):
         index=choice-1

    else:
         print("‚ùå Enter valid Choice\n")
         return

    t=trades[index]

    #Update the P&L
    if t.executed:
        #Take the profit or Loss
        try:
            pnl = int(input("Enter the Profit or Loss: "))

        except ValueError:
            print("‚ùå Invalid input! Please enter a number.")
            return

        t.set_result(pnl)
        print("‚úÖ Trade Updated Successfully\n")

    else:
        print("‚ùå Trade not Executed yet! - EXECUTING...\n")
        t.set_executed()
        print("‚úÖ Trade marked as Executed.\n")

        try:
            pnl = int(input("Enter the Profit or Loss: "))

        except ValueError:
            print("‚ùå Invalid input! Please enter a number.")
            return

        t.set_result(pnl)
        print("‚úÖ Trade Updated Successfully\n")



#Function to check prerequisites
def rule_book():
    while True:
        for p in prerequisites.values():
            response=input(f"{p} (y/n): ").strip().lower()

            if response=="n":
                break

        else:
            print("\n‚úÖ All prerequisites satisfied.\n")
            return
```

<h3>Main menu

```python
#Main Menu Program
def start():
    rule_book()

    while True:
        print("\n1. View Trades")
        print("2. Add Trades")
        print("3. Add Trade P&L")
        print("4. Delete Trades")
        print("5. Calculate P&L")
        print("6. View Dashboard")
        print("7. Mark Executed")
        print("8. Exit\n")

        try:
            choice = int(input("Enter your choice number (1-8): "))
        except ValueError:
            print("‚ùå Invalid input. Enter a number.")
            continue

        match choice:
            case 1:
                print("\n‚û°Ô∏è Viewing Trades")
                view_trades()
            case 2:
                print("\n‚û°Ô∏è Adding Trades")
                add_trade()
            case 3:
                print("\n‚û°Ô∏è Adding Trade P&L")
                update_trade()
            case 4:
                print("\n‚û°Ô∏è Deleting Trades")
                delete_trade()
            case 5:
                print("\n‚û°Ô∏è Calculating P&L")
                calculate_pnl()
            case 6:
                print("\n‚û°Ô∏è Viewing Dashboard")
                view_dashboard()
            case 7:
                print("\n‚û°Ô∏è Marking Executed")
                mark_executed()
            case 8:
                print("\nüî¥ Exiting program. üî¥")
                break
            case _:
                print("\nInvalid choice. Try again.")
```

```python
start()
```
